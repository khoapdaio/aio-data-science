```python
# Global Data
confirmed_df = pd.read_csv(
	'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
deaths_df = pd.read_csv(
	'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
recoveries_df = pd.read_csv(
	'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

```python
confirmed_df = confirmed_df.loc[:, :"8/4/21"]
deaths_df = deaths_df.loc[:, :"8/4/21"]
recoveries_df = recoveries_df.loc[:, :"8/4/21"]
```

#  4. Phân tích Toàn cầu

## Bài toán 1

    Tính tổng số trường hợp mắc, tổng số người chết, tổng số người phục hồi và tổng số đang điều trị theo thời gian (theo ngày).

```python
# Lấy các cột ngày tháng
cols = confirmed_df.keys()

# Lấy dữ liệu từ cột ngày tháng
confirmed = confirmed_df.loc[:, cols[4]:cols[-1]]
deaths = deaths_df.loc[:, cols[4]:cols[-1]]
recoveries = recoveries_df.loc[:, cols[4]:cols[-1]]
dates = confirmed.keys()

# Tính tổng số liệu cho mỗi ngày
total_cases = confirmed.sum(axis = 0)
total_deaths = deaths.sum(axis = 0)
total_recovered = recoveries.sum(axis = 0)
total_active = total_cases - total_deaths - total_recovered

# Tính tỷ lệ tử vong và tỷ lệ phục hồi
mortality_rate = (total_deaths / total_cases) * 100
recovery_rate = (total_recovered / total_cases) * 100

# Chuyển đổi các kết quả về định dạng list để sử dụng hoặc vẽ biểu đồ
total_cases_list = total_cases.tolist()
total_deaths_list = total_deaths.tolist()
total_recovered_list = total_recovered.tolist()
total_active_list = total_active.tolist()
mortality_rate_list = mortality_rate.tolist()
recovery_rate_list = recovery_rate.tolist()

# Định dạng ngày tháng cho dễ hiểu
adjusted_dates = list(dates)
```

```python
#  The number of total cases, total deaths, total recovered, and total active until now.
print("Total confirmed cases: ", total_cases[-1])
print("Total death cases: ", total_deaths[-1])
print("Total recoveries cases: ", total_recovered[-1])
print("Total active cases: ", total_active[-1])
```

## Bài toán 2

    Trực quan hóa số lượng và log(10) tổng số trường hợp, tổng số người chết, tổng số người phục hồi và tổng số đang điều trị theo thời gian (đơn vị ngày) trong 2 biểu đồ (một cho số lượng và một cho log(10)) sử dụng biểu đồ đường.

```python
#Nhập mã lập t# Tạo biểu đồ cho số lượng thực tế
fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (14, 14))
# Biểu đồ cho số lượng thực tế
ax1.plot(adjusted_dates, total_cases_list, label = 'Total Cases')
ax1.plot(adjusted_dates, total_deaths_list, label = 'Total Deaths')
ax1.plot(adjusted_dates, total_recovered_list, label = 'Total Recovered')
ax1.plot(adjusted_dates, total_active_list, label = 'Total Active')
ax1.set_xlabel('Date')
ax1.set_ylabel('Number of Cases')
ax1.set_title('COVID-19 Total Cases Over Time')
ax1.legend()
ax1.grid(True)

# Biểu đồ cho log(10) của số liệu
ax2.plot(adjusted_dates, np.log10(total_cases_list), label = 'Log10 Total Cases')
ax2.plot(adjusted_dates, np.log10(total_deaths_list), label = 'Log10 Total Deaths')
ax2.plot(adjusted_dates, np.log10(total_recovered_list), label = 'Log10 Total Recovered')
ax2.plot(adjusted_dates, np.log10(total_active_list), label = 'Log10 Total Active')
ax2.set_xlabel('Date')
ax2.set_ylabel('Log10 Number of Cases')
ax2.set_title('COVID-19 Log10 Total Cases Over Time')
ax2.legend()
ax2.grid(True)

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()

```

# Bài toán 6:
    Tính tỷ lệ tử vong toàn cầu và tỷ lệ hồi phục theo phần trăm theo thời gian.

```python
#Nhập mã lập trình của bạn tại đây:
# Lấy các cột ngày tháng
cols = confirmed_df.keys()

# Lấy dữ liệu từ cột ngày tháng
confirmed = confirmed_df.loc[:, cols[4]:cols[-1]]
deaths = deaths_df.loc[:, cols[4]:cols[-1]]
recoveries = recoveries_df.loc[:, cols[4]:cols[-1]]
dates = confirmed.keys()

# Tính tổng số liệu cho mỗi ngày
total_cases = confirmed.sum(axis = 0)
total_deaths = deaths.sum(axis = 0)
total_recovered = recoveries.sum(axis = 0)

# Tính tỷ lệ tử vong và tỷ lệ hồi phục theo phần trăm
mortality_rate = (total_deaths / total_cases) * 100
recovery_rate = (total_recovered / total_cases) * 100

# Định dạng ngày tháng cho dễ hiểu
adjusted_dates = pd.to_datetime(dates)

# Trực quan hóa tỷ lệ tử vong và tỷ lệ hồi phục theo thời gian
plt.figure(figsize = (14, 7))

# Biểu đồ cho tỷ lệ tử vong
plt.plot(adjusted_dates, mortality_rate, label = 'Mortality Rate (%)', color = 'red')
plt.plot(adjusted_dates, recovery_rate, label = 'Recovery Rate (%)', color = 'green')
plt.xlabel('Date')
plt.ylabel('Percentage (%)')
plt.title('Global Mortality and Recovery Rates Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

# Bài toán 7
**7.1**

    Trực quan hóa tỷ lệ tử vong theo thời gian (đơn vị ngày) bằng biểu đồ đường.

    Sử dụng đường gạch ngang để trực quan hóa tỷ lệ tử vong trung bình mọi lúc.

```python
#Nhập mã lập trình của bạn tại đây:

# Tính tỷ lệ tử vong trung bình
average_mortality_rate = mortality_rate.mean()

# Định dạng ngày tháng cho dễ hiểu
adjusted_dates = pd.to_datetime(dates)

# Trực quan hóa tỷ lệ tử vong theo thời gian và tỷ lệ tử vong trung bình
plt.figure(figsize = (14, 7))

# Biểu đồ đường cho tỷ lệ tử vong
plt.plot(adjusted_dates, mortality_rate, label = 'Mortality Rate (%)', color = 'red')

# Đường gạch ngang cho tỷ lệ tử vong trung bình
plt.axhline(y = average_mortality_rate, color = 'blue', linestyle = '--',
            label = f'Average Mortality Rate ({average_mortality_rate:.2f}%)')

# Thiết lập biểu đồ
plt.xlabel('Date')
plt.ylabel('Mortality Rate (%)')
plt.title('Global Mortality Rate Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

# Bài toán 9
    Tạo khung dữ liệu country_df với chỉ mục là trường "Country/Region".
    Các thông tin cột này sẽ là:
    - continent: Lục địa của quốc gia.
    - Deaths: Tổng số ca dương tính.
    - Deaths: Tổng số người chết.
    - Recoveries: Tổng số ca hồi phục
    - Active: Tổng số đang điều trị.
    - Mortality Rate: Tỉ lệ tử vong tính theo phần trăm.

```python
#Nhập mã lập trình của bạn tại đây:
# Lấy tổng số các trường hợp, số người chết và số người phục hồi cho từng quốc gia
total_confirmed = confirmed_df.groupby('Country/Region').sum(numeric_only = True).iloc[:, -1]
total_deaths = deaths_df.groupby('Country/Region').sum(numeric_only = True).iloc[:, -1]
total_recoveries = recoveries_df.groupby('Country/Region').sum(numeric_only = True).iloc[:, -1]

# Tính số người đang điều trị (active cases)
total_active = total_confirmed - total_deaths - total_recoveries

# Tính tỷ lệ tử vong (mortality rate)
mortality_rate = (total_deaths / total_confirmed) * 100

# Tạo DataFrame country_df
country_df = pd.DataFrame({
	'Continent': confirmed_df.groupby('Country/Region')['continent'].first(),
	'Total Cases': total_confirmed,
	'Total Deaths': total_deaths,
	'Total Recoveries': total_recoveries,
	'Total Active': total_active,
	'Mortality Rate (%)': mortality_rate
})

country_df.head()
```

# Bài toán 10
    Tạo khung dữ liệu Continental_df với chỉ mục là trường "continent".
    Các thông tin cột này sẽ là:

     - Confirmed: Đất nước lục địa.
     - Deaths: Tổng số người chết.
     - Recoveries: Tổng số lần khôi phục.
     - Active: Tổng số đang điều trị.
     - Mortality Rate: Tỷ lệ tử vong tính theo phần trăm.

```python
#Nhập mã lập trình của bạn tại đây:
# Lấy tổng số các trường hợp, số người chết và số người phục hồi cho từng châu lục
total_confirmed_by_continent = confirmed_df.groupby('continent').sum(numeric_only = True).iloc[:, -1]
total_deaths_by_continent = deaths_df.groupby('continent').sum(numeric_only = True).iloc[:, -1]
total_recoveries_by_continent = recoveries_df.groupby('continent').sum(numeric_only = True).iloc[:, -1]

# Tính số người đang điều trị (active cases) cho từng châu lục
total_active_by_continent = total_confirmed_by_continent - total_deaths_by_continent - total_recoveries_by_continent

# Tính tỷ lệ tử vong (mortality rate) cho từng châu lục
mortality_rate_by_continent = (total_deaths_by_continent / total_confirmed_by_continent) * 100

# Tạo DataFrame continental_df
continental_df = pd.DataFrame({
	'Confirmed': total_confirmed_by_continent,
	'Deaths': total_deaths_by_continent,
	'Recoveries': total_recoveries_by_continent,
	'Active': total_active_by_continent,
	'Mortality Rate (%)': mortality_rate_by_continent
})

continental_df.head()
```

## Biểu đồ đường mẫu
Dưới đây là các hàm mẫu để trực quan hóa các trường hợp sinh động theo thời gian ở mỗi châu lục / quốc gia:

```python
def plot_params(ax, axis_label = None, plt_title = None, label_size = 15, axis_fsize = 15, title_fsize = 20,
                scale = 'linear'):
	# Tick-Parameters
	ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
	ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())
	ax.tick_params(which = 'both', width = 1, labelsize = label_size)
	ax.tick_params(which = 'major', length = 6)
	ax.tick_params(which = 'minor', length = 3, color = '0.8')

	# Grid
	plt.grid(lw = 1, ls = '-', c = "0.7", which = 'major')
	plt.grid(lw = 1, ls = '-', c = "0.9", which = 'minor')

	# Plot Title
	plt.title(plt_title, {'fontsize': title_fsize})

	# Yaxis sacle
	plt.yscale(scale)
	plt.minorticks_on()
	# Plot Axes Labels
	xl = plt.xlabel(axis_label[0], fontsize = axis_fsize)
	yl = plt.ylabel(axis_label[1], fontsize = axis_fsize)


def visualize_covid_cases(confirmed, deaths,
                          continent = None, country = None, state = None, period = None, figure = None,
                          scale = "linear"):
	x = 0
	if figure == None:
		f = plt.figure(figsize = (10, 10))
		# Sub plot
		ax = f.add_subplot(111)
	else:
		f = figure[0]
		# Sub plot
		ax = f.add_subplot(figure[1], figure[2], figure[3])
	ax.set_axisbelow(True)
	plt.tight_layout(pad = 10, w_pad = 5, h_pad = 5)

	stats = [confirmed, deaths]
	label = ["Confirmed", "Deaths"]
	if continent != None:
		params = ["continent", continent]
	elif country != None:
		params = ["country", country]
	else:
		params = ["All", "All"]
	color = ["blue", "red", "green"]
	marker_style = dict(linewidth = 3, linestyle = '-', marker = 'o', markersize = 4, markerfacecolor = '#ffffff')
	for i, stat in enumerate(stats):
		if params[1] == "All":
			cases = np.sum(np.asarray(stat.iloc[:, 5:-1]), axis = 0)[x:]
		else:
			cases = np.sum(np.asarray(stat[stat[params[0]] == params[1]].iloc[:, 5:-1]), axis = 0)[x:]
		date = np.arange(1, cases.shape[0] + 1)[x:]
		plt.plot(date, cases, label = label[i] + " (Total : " + str(cases[-1]) + ")", color = color[i], **marker_style)
		plt.fill_between(date, cases, color = color[i], alpha = 0.3)
	if params[1] == "All":
		Total_confirmed = np.sum(np.asarray(stats[0].iloc[:, 5:-1]), axis = 0)[x:]
		Total_deaths = np.sum(np.asarray(stats[1].iloc[:, 5:-1]), axis = 0)[x:]
	else:
		Total_confirmed = np.sum(np.asarray(stats[0][stats[0][params[0]] == params[1]].iloc[:, 5:-1]), axis = 0)[x:]
		Total_deaths = np.sum(np.asarray(stats[1][stats[1][params[0]] == params[1]].iloc[:, 5:-1]), axis = 0)[x:]

	text = stats[0].columns[5] + " to " + stats[0].columns[-1] + "\n"
	text += "Mortality rate : " + str(int(Total_deaths[-1] / (Total_confirmed[-1]) * 10000) / 100) + "\n"
	text += "Last 5 Days:\n"
	text += "Confirmed : " + str(Total_confirmed[-1] - Total_confirmed[-6]) + "\n"
	text += "Deaths : " + str(Total_deaths[-1] - Total_deaths[-6]) + "\n"
	text += "Last 24 Hours:\n"
	text += "Confirmed : " + str(Total_confirmed[-1] - Total_confirmed[-2]) + "\n"
	text += "Deaths : " + str(Total_deaths[-1] - Total_deaths[-2]) + "\n"

	plt.text(0.02, 0.78, text, fontsize = 15, horizontalalignment = 'left', verticalalignment = 'top',
	         transform = ax.transAxes, bbox = dict(facecolor = 'white', alpha = 0.4))

	# Plot Axes Labels
	axis_label = ["Days (" + confirmed_df.columns[5] + " - " + confirmed_df.columns[-1] + ")", "No of Cases"]

	# Plot Parameters
	plot_params(ax, axis_label, scale = scale)

	# Plot Title
	if params[1] == "All":
		plt.title("COVID-19 Cases World", {'fontsize': 25})
	else:
		plt.title("COVID-19: " + params[1], {'fontsize': 25})

	# Legend Location
	l = plt.legend(loc = "best", fontsize = 15)

	if figure == None:
		plt.show()


```
