#### TASK: Mở ảnh *dog_backpack.jpg* từ folder DATA và hiển thị trong notebook. Đảm bảo sửa theo thứ tự RGB.

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Load the image (replace 'dog.jpg' with the path to your image)
image_path = '../DATA/dog_backpack.jpg'
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
```

#### TASK: Lật ngược ảnh và hiển thị trong notebook.

```python
plt.imshow(cv2.flip(image_rgb, 0))
```

#### TASK: Vẽ một hình chữ nhật trống MÀU ĐỎ quanh khuôn mặt của chú chó và hiển thị trong notebook.

```python
# Define the rectangle coordinates (replace with actual face coordinates)
x1, y1 = 200, 360  # Top-left corner
x2, y2 = x1 + 400, y1 + 360  # Bottom-right corner

# Draw the rectangle on the image
color = (255, 0, 0)  # Red color in BGR
thickness = 10  # Thickness of the rectangle border
image_with_rectangle = cv2.rectangle(image_rgb.copy(), (x1, y1), (x2, y2), color, thickness)

# Display the image with matplotlib
plt.figure(figsize = (4, 6))
plt.imshow(image_with_rectangle)
```

#### TASK: Vẽ một HÌNH TAM GIÁC MÀU XANH DƯƠNG ở giữa ảnh. Có thể chọn kích thước và góc tùy ý, nhưng phải là hình tam giác (3 cạnh) theo bất kỳ hướng nào.

```python
# Định nghĩa ba đỉnh của tam giác
pt1 = (450, 200)  # Điểm 1
pt2 = (150, 700)  # Điểm 2
pt3 = (750, 700)  # Điểm 3

# Draw the rectangle on the image
color = (0, 0, 255)  # Red color in BGR
triangle_cnt = np.array([pt1, pt2, pt3])
image_with_triangle = cv2.drawContours(image_rgb.copy(), [triangle_cnt], 0, color, thickness = 10)

# Display the image with matplotlib
plt.figure(figsize = (4, 6))
plt.imshow(image_with_triangle)
```

### BONUS TASK. Bạn có thể tô màu tam giác này không? Điều này cần một hàm khác, không có trong bài giảng. Hãy thử xem liệu bạn có thể tìm thấy trên Google không.

[CLICK VÀO LIÊN KẾT SAU ĐỂ XEM GỢI Ý](https://docs.opencv.org/3.0-beta/modules/imgproc/doc/drawing_functions.html#fillpoly)

```python
# Định nghĩa ba đỉnh của tam giác
pt1 = (450, 200)  # Điểm 1
pt2 = (150, 700)  # Điểm 2
pt3 = (750, 700)  # Điểm 3

# Draw the rectangle on the image
color = (0, 0, 255)  # Red color in BGR
triangle_cnt = np.array([pt1, pt2, pt3])
image_with_triangle_fill = cv2.fillPoly(image_rgb.copy(), [triangle_cnt], color)

# Display the image with matplotlib
plt.figure(figsize = (4, 6))
plt.imshow(image_with_triangle_fill)
```
